```python
set=40

datasets, labels, train = facedataset(set)

X_mean = np.array(train).mean(axis=0)
X_std = np.array(train).std(axis=0)
train_norm = (train - X_mean)/X_std
datasets_norm = (datasets - X_mean)/X_std

trainnew = []
trainnew_n=[]
trainnew_n.append(train_norm)
trainnew.append(train)
X=[]
H = []

for i in range(9):
    #print('i', i+1)
    if i!=0:
        traintmp =[]
        datasets_norm_tmp =[]
        for l in range(len(datasets_norm)):
            traintmp.append(datasets_norm[l][0])
            datasets_norm_tmp.append(np.delete(datasets_norm[l],0,0))
        datasets_norm = np.array(datasets_norm_tmp)
    #print(datasets_norm.shape)
        trainnew_n.append(traintmp)

        print(len(trainnew_n[0])*len(trainnew_n))
        X.append(len(trainnew_n[0])*len(trainnew_n))

        traintmp = []

        for l in range(len(datasets)):
            traintmp.append(datasets[l][0])
            datasets[l].pop(0)

        trainnew.append(traintmp)
    else:
        print(set)
        X.append(40)
    #считаем зависимость
    label = 0
    loss = 0
    l_all = 0

    for dataset in datasets:
        loss = 0
        for img in dataset:
            all_dist_H = []
            for train in trainnew:
                dist = [H_clsf(img, img_c) for img_c in train]
                dist = dist / np.sum(dist)
                #print(dist.argmin())
                all_dist_H.append(dist.argmin())
            tmp_labels = np.array([all_dist_H.count(g) for g in all_dist_H])
            loss = loss + 1 if all_dist_H[tmp_labels.argmax()] == label else loss

            #l = l + 1 if np.mean(all_dist, axis=0).argmin() == label else l
           # print(all_dist)
        l_all += loss / len(dataset)
        label += 1
    print('Brightness histogram accuracy:', l_all / len(datasets))
    H.append(l_all / len(datasets))
    
plt.plot(X, H)
plt.title("Зависимость точности от числа тестовых изображений для Brightness histogram")
plt.xlabel('Кол-во тестовых изображений')
plt.ylabel('Точность')
plt.show()
```
